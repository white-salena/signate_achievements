```python
# 方角を重みづけする
import pandas as pd
import numpy as np
import re
```

```python
train1 = pd.read_csv('train.csv')
test1 = pd.read_csv('test.csv')
```

```python
train = train1[['id', '方角']].copy()
test = test1[['id', '方角']].copy()
```

```python
# 変換ルールの定義
conversion_rules = {
    '東': 1.5,
    '西': 1,
    '南': 2.5,
    '北': 0.5
}
```

```python
# 方角を数値に変換する関数
def convert_direction(direction):
    # NaNを処理
    if pd.isna(direction) or direction == '':
        return 0
    elif direction in conversion_rules:
        return conversion_rules[direction]
    else:
        # 方角が混合の場合（例: 南東, 北西など）を処理
        value = 0
        if '南' in direction:
            value += conversion_rules['南']
        if '東' in direction:
            value += conversion_rules['東']
        if '西' in direction:
            value += conversion_rules['西']
        if '北' in direction:
            value += conversion_rules['北']
        return value
```

```python
# 方角の変換を適用
train['方角'] = train['方角'].apply(convert_direction)
test['方角'] = test['方角'].apply(convert_direction)
```

```python
# 全体の平均値出すために一回結合
df = pd.concat([train, test], ignore_index=True)
```

```python
# 方角が0の行を平均値で補完
mean_value = df['方角'].replace(0, np.nan).mean()  # 0をNaNに置き換えた後に平均を計算
train['方角'] = train['方角'].replace(0, mean_value)  # 0を平均値で補完
test['方角'] = test['方角'].replace(0, mean_value)  # 0を平均値で補完
```

```python
# 方角を小数点第1位までで揃える
train['方角'] = train['方角'].round(1)
test['方角'] = test['方角'].round(1)
```

```python
train.to_csv('train_direction.csv', index=False)
test.to_csv('test_direction.csv', index=False)
```
