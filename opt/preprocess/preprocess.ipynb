```python
import sys
import os

# 1つ上のディレクトリにパスを通す
sys.path.append(os.path.abspath('../'))

from module.Dataset import read_train_tsv, to_tsv_preprocess
from module.Feature import UserFeature, ItemFeature, compute_relevance
```

```python
# 学習データ読み込み
train_A = read_train_tsv('A')
train_B = read_train_tsv('B')
train_C = read_train_tsv('C')
train_D = read_train_tsv('D')
```

```python
# 関連度スコアの設定
train_A['relevance'] = train_A.apply(compute_relevance, axis=1)
train_B['relevance'] = train_B.apply(compute_relevance, axis=1)
train_C['relevance'] = train_C.apply(compute_relevance, axis=1)
train_D['relevance'] = train_D.apply(compute_relevance, axis=1)
```

```python
# ユーザ関連の特徴量生成クラスをインスタンス化
user_A = UserFeature(train_A, 'relevance')
user_B = UserFeature(train_B, 'relevance')
user_C = UserFeature(train_C, 'relevance')
user_D = UserFeature(train_D, 'relevance')
```

```python
# ユーザID毎に行動種別の集計(但し関連度を基準)
user_A.relevance_count()
user_B.relevance_count()
user_C.relevance_count()
user_D.relevance_count()
```

```python
# ユーザID毎に曜日単位で行動種別の集計(但し関連度を基準)
user_A.weekday_count()
user_B.weekday_count()
user_C.weekday_count()
user_D.weekday_count()
```

```python
# 商品関連の特徴量生成クラスをインスタンス化
item_A = ItemFeature(user_A.df, 'relevance')
item_B = ItemFeature(user_B.df, 'relevance')
item_C = ItemFeature(user_C.df, 'relevance')
item_D = ItemFeature(user_D.df, 'relevance')
```

```python
# 商品ID毎に行動種別の集計(但し関連度を基準)
item_A.relevance_count()
item_B.relevance_count()
item_C.relevance_count()
item_D.relevance_count()
```
