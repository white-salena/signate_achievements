```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# データの読み込み
df = pd.read_csv("train_balanced_pre.csv")

# 正規化したい列（IDと賃料を除く）
columns_to_normalize = df.columns.difference(['id', '賃料'])

# MinMaxScalerのインスタンスを作成
scaler = MinMaxScaler()

# 指定した列を正規化
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

# 正規化されたデータフレームを表示
print(df)
```
