```python
# 間取りから部屋数を割り出す
import pandas as pd
import re
```

```python
train1 = pd.read_csv('train.csv')
test1 = pd.read_csv('test.csv')
```

```python
train = train1[['id', '間取り']].copy()
test = test1[['id', '間取り']].copy()
```

```python
# 変換ルールの定義
conversion_rules = {
    'R': 0,
    'K': 0,
    'DK': 1,
    'LK': 1,
    'LDK': 1.5,
    '+S': 1
}
```

```python
# 間取りを数値に変換する関数
def convert_room(val):
    # 最初の数値を取得
    match = re.match(r'(\d+)', val)
    if not match:
        return None  # 数値が見つからない場合はNoneを返す
    base_value = int(match.group(1))  # 数値部分
    
    # 間取り部分を取得（例: LDK, DK, +Sなど）
    room_type = re.findall(r'[A-Z]+|\+S', val)
    
    # 間取りに基づいて数値を変換し、合計を計算
    additional_value = sum(conversion_rules.get(rt, 0) for rt in room_type)
    
    return base_value + additional_value
```

```python
# '間取り'列を数値に変換して'部屋数'に格納
train['部屋数'] = train['間取り'].apply(convert_room)
test['部屋数'] = test['間取り'].apply(convert_room)
```

```python
train.drop(columns=['間取り'], inplace=True)
test.drop(columns=['間取り'], inplace=True)
```

```python
train.to_csv('train_rooms.csv', index=False)
test.to_csv('test_rooms.csv', index=False)
```
