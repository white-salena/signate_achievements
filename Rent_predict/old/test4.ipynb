```python
import pandas as pd

# データの作成
data = {
    'id': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    '賃料': [75000, 76000, 110000, 150000, 74000, 160000, 78500, 115000, 184327],
    '所在地': ['北区', '中央区', '渋谷区', '杉並区', '葛飾区', '荒川区', '練馬区', '目黒区', '文京区']
}

train_df = pd.DataFrame(data)

# 所在地を数値に変換するマッピング
location_mapping = {location: idx for idx, location in enumerate(train_df['所在地'].unique())}

# 所在地列を数値に変換
train_df['所在地'] = train_df['所在地'].map(location_mapping)

# 結果の表示
print(train_df)

# テストデータの作成
test_data = {
    'id': [10, 11, 12],
    '賃料': [80000, 90000, 120000],
    '所在地': ['中央区', '文京区', '渋谷区']  # 並び順が異なる
}

test_df = pd.DataFrame(test_data)

# テストデータの所在地も数値に変換
test_df['所在地'] = test_df['所在地'].map(location_mapping)

# テストデータの結果の表示
print(test_df)
```
