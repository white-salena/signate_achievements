```python
import pandas as pd
import pickle
```

```python
test = pd.read_csv("test_pre.csv")
```

```python
#モデルの読み込み
with open("model_XGB.pickle", "rb") as fp:
    model = pickle.load(fp)
```

```python
predict_y = model.predict(test)
```

```python
test["y_pre"] = predict_y
```

```python
test['y'] = test['面積'] * test['y_pre']
```

```python
test['y'] = test['y'].round().astype(int)
```

```python
inference = test[['id', 'y']]
```

```python
inference.to_csv('inference_XGB.csv', index=False, header=None)
```
