# 学習用データの説明

train_A.tsv - 人材<br>
train_B.tsv - 旅行<br>
train_C.tsv - 不動産<br>
train_D.tsv - アパレル<br>

# カラムの説明

0 - user_id    - ユーザーID<br>
1 - product_id - 商品ID<br>
2 - event_type - 行動種別<br>
3 - ad         - コンバージョンが広告経由かどうか (コンバージョン以外は-1)<br>
4 - time_stamp - タイムスタンプ<br>

# 行動種別(event_type)の意味

3 - cv - コンバージョンの略で商品を購入すること<br>
2 - cl - クリックの略で広告にクリックすること<br>
1 - pd - 商品の詳細ページを閲覧すること<br>
0 - ca - 商品をカートに入れること
