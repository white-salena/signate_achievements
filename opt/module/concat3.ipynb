```python
import pandas as pd
```

```python
pred_A = pd.read_csv('../dataset/predict3_A.tsv', sep="\t")
pred_B = pd.read_csv('../dataset/predict3_B.tsv', sep="\t")
pred_C = pd.read_csv('../dataset/predict3_C.tsv', sep="\t")
pred_D = pd.read_csv('../dataset/predict3_D.tsv', sep="\t")
```
