```python
import pandas as pd
```

```python
pred_A = pd.read_csv('./predict/predictions_A.tsv', sep="\t")
pred_B = pd.read_csv('./predict/predictions_B.tsv', sep="\t")
pred_C = pd.read_csv('./predict/predictions_C.tsv', sep="\t")
pred_D = pd.read_csv('./predict/predictions_D.tsv', sep="\t")
```

```python
pred = pd.concat([pred_A, pred_B, pred_C, pred_D], ignore_index=True)
```

```python
pred.to_csv('./predict/predictions.tsv', sep='\t', index=False, header=False)
```
