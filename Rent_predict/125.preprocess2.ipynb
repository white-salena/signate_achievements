```python
# 1部屋辺りの面積を割り出す
import pandas as pd
```

```python
train1 = pd.read_csv('train.csv')
test1 = pd.read_csv('test.csv')
```

```python
train = train1[['id', '面積']].copy()
test = test1[['id', '面積']].copy()
```

```python
train2 = pd.read_csv('train_rooms.csv')
test2 = pd.read_csv('test_rooms.csv')
```

```python
train['部屋数'] = train2['部屋数'].copy()
test['部屋数'] = test2['部屋数'].copy()
```

```python
# m2を削除して数値に変換
train['面積'] = train['面積'].str.replace('m2', '').astype(float)
test['面積'] = test['面積'].str.replace('m2', '').astype(float)
```

```python
train['部屋面積'] = train['面積'] / train['部屋数']
test['部屋面積'] = test['面積'] / test['部屋数']
```

```python
# 部屋面積を小数点第2位までで揃える
train['部屋面積'] = train['部屋面積'].round(2)
test['部屋面積'] = test['部屋面積'].round(2)
```

```python
train.drop(columns=['面積', '部屋数'], inplace=True)
test.drop(columns=['面積', '部屋数'], inplace=True)
```

```python
train.to_csv('train_area.csv', index=False)
test.to_csv('test_area.csv', index=False)
```
