```python
import pandas as pd
import pickle
```

```python
test = pd.read_csv("test1.csv")
test2 = pd.read_csv("test.csv")
```

```python
#モデルの読み込み
with open("model_RFR.pickle", "rb") as fp:
    model = pickle.load(fp)
```

```python
predict_y = model.predict(test)
```

```python
inference = pd.DataFrame()
inference['id'] = test2['id']
```

```python
inference["y"] = predict_y
```

```python
inference.head()
```

```python
inference.to_csv('inference_RFR.csv', index=False, header=None)
```
