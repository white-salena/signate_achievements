```python
import pandas as pd
```

```python
# 色々加工したデータ読み込み
train = pd.read_csv('train3.csv')
test = pd.read_csv('test3.csv')
```

```python
# アクセスの加工済みデータ読み込み
train_accese = pd.read_csv('train_accese_nearest1.csv')
test_accese = pd.read_csv('test_accese_nearest1.csv')
```

```python
# id列はいらないから削除
train_accese = train_accese.drop(columns=['id'])
test_accese = test_accese.drop(columns=['id'])
```

```python
# 間取りの加工済みデータ読み込み
train_rooms = pd.read_csv('train_rooms.csv')
test_rooms = pd.read_csv('test_rooms.csv')
```

```python
# id列はいらないから削除
train_rooms = train_rooms.drop(columns=['id'])
test_rooms = test_rooms.drop(columns=['id'])
```

```python
# 所在階の加工済みデータ読み込み
train_floor = pd.read_csv('train_floor.csv')
test_floor = pd.read_csv('test_floor.csv')
```

```python
# id列はいらないから削除
train_floor = train_floor.drop(columns=['id'])
test_floor = test_floor.drop(columns=['id'])
```

```python
# 面積の加工済みデータ読み込み
train_area = pd.read_csv('train_area.csv')
test_area = pd.read_csv('test_area.csv')
```

```python
# id列はいらないから削除
train_area = train_area.drop(columns=['id'])
test_area = test_area.drop(columns=['id'])
```

```python
# 築年数の加工済みデータ読み込み
train_year = pd.read_csv('train_year.csv')
test_year = pd.read_csv('test_year.csv')
```

```python
# id列はいらないから削除
train_year = train_year.drop(columns=['id'])
test_year = test_year.drop(columns=['id'])
```

```python
# 方角の加工済みデータ読み込み
train_direction = pd.read_csv('train_direction.csv')
test_direction = pd.read_csv('test_direction.csv')
```

```python
# id列はいらないから削除
train_direction = train_direction.drop(columns=['id'])
test_direction = test_direction.drop(columns=['id'])
```

```python
train = pd.concat([train, train_accese], axis=1)
test = pd.concat([test, test_accese], axis=1)
```

```python
train = pd.concat([train, train_rooms], axis=1)
test = pd.concat([test, test_rooms], axis=1)
```

```python
train = pd.concat([train, train_floor], axis=1)
test = pd.concat([test, test_floor], axis=1)
```

```python
train = pd.concat([train, train_area], axis=1)
test = pd.concat([test, test_area], axis=1)
```

```python
train = pd.concat([train, train_year], axis=1)
test = pd.concat([test, test_year], axis=1)
```

```python
train = pd.concat([train, train_direction], axis=1)
test = pd.concat([test, test_direction], axis=1)
```

```python
train.to_csv('train_pre.csv', index=False)
test.to_csv('test_pre.csv', index=False)
```
