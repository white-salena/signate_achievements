```python
# 分割したアクセス列の整備
import pandas as pd
```

```python
# CSVファイルを読み込み
df = pd.read_csv('train_accese.csv')
```

```python
# 徒歩分数を数値に変換（エラーが出る場合はNoneに置き換え）
df['徒歩1_分'] = pd.to_numeric(df['徒歩1_分'], errors='coerce')
df['徒歩2_分'] = pd.to_numeric(df['徒歩2_分'], errors='coerce')
df['徒歩3_分'] = pd.to_numeric(df['徒歩3_分'], errors='coerce')
```

```python
# 最も徒歩分数が少ない駅の路線名と駅名を選択する関数
def select_nearest_station(row):
    # 徒歩分数をリストにして最小値を選ぶ
    min_walk_time = min(row['徒歩1_分'], row['徒歩2_分'], row['徒歩3_分'])
    
    # 最小値に対応する駅と路線を選択
    if min_walk_time == row['徒歩1_分']:
        return row['路線1_名'], row['駅1_名'], row['徒歩1_分']
    elif min_walk_time == row['徒歩2_分']:
        return row['路線2_名'], row['駅2_名'], row['徒歩2_分']
    else:
        return row['路線3_名'], row['駅3_名'], row['徒歩3_分']
```

```python
# 新しい列に最寄り駅情報を格納
df[['最寄り路線', '最寄り駅', '所要時間']] = df.apply(select_nearest_station, axis=1, result_type='expand')
```

```python
# 必要な列だけを表示
df_nearest = df[['id', '最寄り路線', '最寄り駅', '所要時間']]
```

```python
# 整数に丸める
df_nearest['所要時間'] = df_nearest['所要時間'].round().astype(int)
```

```python
df_nearest.to_csv('train_accese_nearest.csv', index=False)
```
