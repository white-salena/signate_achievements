```python
# 最寄り路線と最寄り駅を数値変換
import pandas as pd
```

```python
train = pd.read_csv('train_accese_nearest.csv')
test = pd.read_csv('test_accese_nearest.csv')
```

```python
# マッピング用に一回結合
df = pd.concat([train, test], ignore_index=True)
```

```python
# 最寄り路線を数値に変換するマッピング
location_mapping1 = {location: idx for idx, location in enumerate(df['最寄り路線'].unique())}
```

```python
# 最寄り路線を数値に変換
train['最寄り路線'] = train['最寄り路線'].map(location_mapping1)
test['最寄り路線'] = test['最寄り路線'].map(location_mapping1)
```

```python
# 最寄り駅を数値に変換するマッピング
location_mapping2 = {location: idx for idx, location in enumerate(df['最寄り駅'].unique())}
```

```python
# 最寄り駅を数値に変換
train['最寄り駅'] = train['最寄り駅'].map(location_mapping2)
test['最寄り駅'] = test['最寄り駅'].map(location_mapping2)
```

```python
train.to_csv('train_accese_nearest1.csv', index=False)
test.to_csv('test_accese_nearest1.csv', index=False)
```
