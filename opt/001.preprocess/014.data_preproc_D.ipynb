```python
import pandas as pd
```

```python
#ファイル読み込み
df = pd.read_csv('../000.data/train/train_D.tsv', sep="\t")
```

```python
#重複行の削除
df_unique = df.drop_duplicates()
```

```python
#各列の情報確認
df_unique.info()
```

```python
#行動種別(event_type)毎の数
df_unique['event_type'].value_counts()
```

```python
#コンバージョンの数
df_unique['ad'].value_counts()
```

```python
#データフレームのコピー
train = df_unique.copy()
```

```python
#ユーザー毎の人気度を追加
train['user_interest'] = train[train['event_type'].isin([0, 1, 3])].groupby('user_id')['event_type'].transform('count')
```

```python
#商品毎の人気度を追加
train['product_interest'] = train[train['event_type'].isin([0, 1, 3])].groupby('product_id')['event_type'].transform('count')
```

```python
#user_idとproduct_idを数値化
train['user_id'] = train['user_id'].astype('category').cat.codes
train['product_id'] = train['product_id'].astype('category').cat.codes
```

```python
# time_stamp を datetime 型に変換
train['time_stamp'] = pd.to_datetime(train['time_stamp'])
```
