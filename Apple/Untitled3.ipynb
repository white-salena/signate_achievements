```python
import xgboost as xgb
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
import joblib

# データを読み込む（ここでは仮のデータフレームを作成します）
# 実際にはあなたのデータを読み込んでください
X = pd.DataFrame(np.random.rand(100, 10))  # 特徴量データ
y = pd.Series(np.random.rand(100))         # 目的変数データ

# パラメータ設定
params = {
    'objective': 'reg:squarederror',
    'learning_rate': 0.1,
    'max_depth': 3,
    'n_estimators': 100
}

# KFoldの設定
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# モデルのリストを保持するリスト
models = []

# それぞれの分割についてモデルを学習
for train_index, val_index in kf.split(X):
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]
    
    # モデルのインスタンスを作成
    model = xgb.XGBRegressor(**params)
    
    # モデルの学習
    model.fit(X_train, y_train)
    
    # 学習済みモデルをリストに追加
    models.append(model)

# モデルを保存
joblib.dump(models, 'xgb_kfold_models.pkl')

# 保存されたモデルをロードして予測を行う関数を作成
def predict_with_models(models, X):
    preds = np.zeros((len(models), X.shape[0]))
    for i, model in enumerate(models):
        preds[i, :] = model.predict(X)
    return preds.mean(axis=0)

# モデルをロード
loaded_models = joblib.load('xgb_kfold_models.pkl')

# 新しいデータに対して予測を行う（ここでは仮の新しいデータを使用）
X_new = pd.DataFrame(np.random.rand(10, 10))

# 予測を実施
predictions = predict_with_models(loaded_models, X_new)

print(predictions)
```
